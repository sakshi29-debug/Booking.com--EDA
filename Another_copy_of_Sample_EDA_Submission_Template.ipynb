## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
import pandas as pd
import numpy as np

# 1. Load the dataset
df = pd.read_csv('Hotel Bookings.csv')  # Replace with your file name

# 2. View basic structure
print("Initial shape:", df.shape)
print(df.info())
print(df.head())

# 3. Rename columns for consistency (optional)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# 4. Handle missing values
# Fill missing numerical values with median
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill missing categorical values with mode or 'Unknown'
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0] if not df[col].mode().empty else 'Unknown')

# 5. Remove duplicates
df = df.drop_duplicates()
````

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code
hotel_counts = df['hotel'].value_counts()

#Plot pie chart
plt.figure(figsize=(6, 4))
plt.pie(
    hotel_counts,
    labels=hotel_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66b3ff', '#99ff99']
)
plt.title('Booking Distribution by Hotel Type (Pie Chart)')
plt.axis('equal')  # Ensures it's a perfect circle
plt.show()
````

##### 1. Why did you pick the specific chart?

The goal of our code is to visualize the distribution of bookings across hotel types, which makes a pie chart a natural and intuitive choice.
A pie chart is ideal when you're trying to show how a whole is divided into parts.
Viewers can quickly grasp that, for example, “City Hotel has 61.1% of all bookings” just by glancing


##### 2. What is/are the insight(s) found from the chart?

1. City Hotels are More Popular
A majority of the bookings are for City Hotels.

This suggests that travelers may prefer city locations, possibly due to business trips, better connectivity, or urban attractions.

2. Resort Hotels Still Hold a Significant Share
35% of bookings are for Resort Hotels, indicating strong demand for leisure or vacation stays.

Shows a healthy balance between business and leisure travelers.

3. Travel Purpose May Influence Hotel Choice
The high percentage of City Hotel bookings may reflect short stays, business travel, or transit visits.

Resort Hotels may attract longer stays or family vacations.

4. Marketing Strategy Opportunity
If you’re a hotel chain or OTA (online travel agency), consider targeting City Hotel guests with upselling or cross-promotions.

Conversely, promote Resort Hotels during holiday seasons or with family packages to boost their share.



##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, the insights from the pie chart can lead to positive business impact through smarter marketing, resource planning, and customer segmentation.

However, ignoring the performance gap in Resort Hotels could lead to negative growth or missed revenue opportunities — especially in leisure and seasonal markets.

#### Chart - 2

````python
# Chart - 2 visualization code
# 2. Booking cancellations by hotel type
plt.figure(figsize=(6, 4))
sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Set2')
plt.title('Booking Cancellation by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Count')
plt.legend(title='Is Cancelled')
plt.show()
````

##### 1. Why did you pick the specific chart?

1. Clear Comparison of Counts
The countplot directly shows the number of bookings for each hotel type.

Using hue='is_canceled' splits each hotel’s bookings into cancelled vs. not cancelled, making it easy to compare the two categories side by side.

2. Visualizing Distribution & Proportions Together
You get to see both total booking volumes (bar height) and cancellation distribution (color split) simultaneously.

This helps identify:

Which hotel type has more bookings overall

Which hotel type experiences more cancellations (both absolute and relative)

3. Categorical Data Handling
Both hotel and is_canceled are categorical variables.

Countplots are ideal for counting categorical occurrences and are more intuitive than pie charts or stacked bar charts for this kind of dual-category data.

4. Easy to Interpret & Compare
Bar charts are a widely understood visualization.

Showing two categories side by side helps spot patterns like:

“City Hotels have higher cancellations”

“Resort Hotels have fewer cancellations but also fewer bookings”

##### 2. What is/are the insight(s) found from the chart?

1. City Hotels Have More Cancellations
The bar for canceled bookings is significantly higher for City Hotels than for Resort Hotels.

This suggests higher cancellation rates for City Hotels.

🧠 Reason: Business travelers who often book City Hotels may cancel due to last-minute plan changes or flexible bookings.

2. City Hotels Have Higher Total Bookings
Both canceled and non-canceled bookings are higher for City Hotels overall.

This means City Hotels are more in demand, possibly due to:

Urban location

Proximity to business hubs

Easier accessibility

3. Resort Hotels Have Fewer Cancellations
Resort Hotel bookings are more stable and less frequently canceled.

Guests booking resorts are likely planning vacations, and thus less likely to cancel.

🧠 Business Implication: Resort bookings are more reliable for revenue forecasting.

4. Potential for Policy Adjustment
The high cancellation rate for City Hotels could lead to:

Lost revenue

Unused inventory

Increased operational uncertainty

✅ Actionable Insight: Consider non-refundable or semi-flexible rate plans for City Hotels to reduce cancellation volume.

#### Chart - 3

````python
# Chart - 3 visualization code
# 3. Average Daily Rate (ADR) by hotel type
plt.figure(figsize=(6, 4))
sns.boxplot(x='hotel', y='adr', data=df)
plt.title('ADR Distribution by Hotel Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.xlabel('Hotel')
plt.ylim(0, 500)  # Optional: adjust for better visualization
plt.show()
````

##### 1. Why did you pick the specific chart?

The boxplot gives a statistical summary: median, quartiles, and outliers of ADR.

Helps you quickly understand:

Which hotel type charges more on average

How spread out or consistent prices are

Presence of extreme values (e.g., holiday spikes, luxury bookings)

##### 2. What is/are the insight(s) found from the chart?

1. Which Hotel Type Has Higher Median ADR
The central line inside the box represents the median ADR.

If City Hotels have a higher median, they generally earn more per night.

2. Price Variability
A wider box or longer whiskers indicate more variability in ADR.

Resort Hotels often have more seasonal or luxury pricing, so greater ADR spread is common.

3. Outliers Detection
Individual dots outside the whiskers show outliers—very high prices.

These might represent special events, suites, or pricing anomalies.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

✅ Positive Insights:
If Resort Hotels have higher ADR, they can be targeted for revenue growth during holidays.

If City Hotels have consistent ADR, they are good for steady income forecasting.

⚠️ Risks (if unaddressed):
High ADR outliers may discourage price-sensitive customers if not managed carefully.

Low ADR during high demand periods could indicate missed revenue opportunities.



#### Chart - 4

````python
# Chart - 4 visualization code
# 4. Monthly booking trend
monthly_bookings = df['arrival_date_month'].value_counts().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])

plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_bookings.index, y=monthly_bookings.values, palette='coolwarm')
plt.title('Monthly Booking Trends')
plt.xticks(rotation=45)
plt.ylabel('Number of Bookings')
plt.xlabel('Month')
plt.show()
````

##### 1. Why did you pick the specific chart?

It visualizes how bookings vary by month, revealing:

Peak and off-peak seasons

Customer behavior patterns (e.g. holiday travel spikes)

Planning opportunities for marketing, staffing, and pricing



##### 2. What is/are the insight(s) found from the chart?

1. Peak Booking Months
Months with the highest bars (often July, August) indicate high demand seasons.

This is usually due to summer vacations, public holidays, or tourist seasons.

2. Off-Peak Months
Months with lower bookings (e.g., November, January) may reflect:

Post-holiday slowdown

School or work cycles

Less favorable weather in resort areas

3. Marketing & Revenue Strategy
You can schedule discounts/promos in off-peak months to drive demand.

Apply surge pricing in high-demand months to maximize revenue.

Business travelers (1–2 nights)

Weekend leisure trips

2. Few Long Stays
As the number of nights increases, the bars shrink—indicating fewer long-term stays.

A small number of guests stay for extended durations (10+ nights), possibly:

Vacationers

Long-term corporate stays

Relocations

3. Outliers / Unusual Durations
If you spot isolated bars beyond ~20 nights, they could be:

Booking anomalies
